🧠 Problema: O Jantar dos Filósofos
Cinco filósofos estão sentados ao redor de uma mesa circular. Cada filósofo pensa indefinidamente, mas ocasionalmente fica com fome e tenta comer. No centro da mesa há um prato de comida, e entre cada par de filósofos há um único garfo (ou talher). Para comer, um filósofo precisa pegar os dois garfos ao seu lado: o da esquerda e o da direita. Após comer, ele coloca os garfos de volta e volta a pensar.

Você deve implementar uma simulação em Python que modele esse problema de forma concorrente, usando threads e mecanismos de sincronização como locks ou semáforos, de forma que:

- Nenhum filósofo passe fome (evite deadlocks e starvation).

- Filósofos possam alternar entre pensar e comer.

- O programa possa rodar por um tempo determinado ou até interrupção manual, mostrando no console os estados de cada filósofo (pensando, com fome, comendo).

Requisitos Técnicos:
- Use a biblioteca threading ou Multiprocessing.

- Cada filósofo deve ser uma thread.

- Use threading.Lock ou threading.Semaphore para representar os garfos.

- Implemente uma forma segura de controle para evitar condições de corrida e deadlocks.


Mostre no console mensagens como:
"Filósofo 3 está pensando."
"Filósofo 3 está com fome."
"Filósofo 3 está comendo."
